```python
#Importing libraries
import seaborn as sns
import pandas as pd
import numpy as np
import scipy.stats
```

```python
kashti=sns.load_dataset("titanic")
kashti.head()
```

```python
np.cov(kashti["age"],kashti["fare"])  # Covariance
```

```python
# find out covariance with respect to columns

cov_mat=np.stack((kashti["age"],kashti["fare"]), axis=0)
cov_mat
# print(np.cov(cov_mat))
```

# Its difficult to interpret covariance thats whay

```python
kashti.corr() #simple corr
```

```python
kashti.corr(method="pearson")  # for gaussian data
```

```python
kashti.corr(method="spearman") # for non gaussian data
```

```python
sns.regplot(kashti["fare"],kashti["survived"], data=kashti)
```

```python
# iris data
phool=sns.load_dataset("iris")
phool.head()
```

```python
phool.corr() #simple corr
```

```python
# positive corr
sns.regplot(phool["sepal_length"],phool["petal_width"], data=kashti)
```

```python
# negative corr
sns.regplot(phool["sepal_width"],phool["petal_length"], data=phool)
```

```python
# neutral corr
sns.regplot(phool["petal_length"],phool["petal_length"], data=phool)
```

```python
# corr on heatmap
sns.heatmap(kashti.corr())
```

```python
# corr on heatmap with annotation
sns.heatmap(kashti.corr(), annot=True)
```

```python
# with style
kashti.corr().style.background_gradient(cmap="coolwarm")
```

```python
# pairplot 
sns.pairplot(phool)
```
