# 3.Find missing values

```python
# yahan hm missing values dekh sakte hain aur agr 
# ziada hon to remove kr dete hain
df.isnull().sum()
```

```python
# missing values ke percentage 
# df.isnull().sum()/df.shape[0]*100
ks.isnull().sum()/ks.shape[0]*100 # ks.isnull().sum()/891*100
```

```python
# p.column ke missing values ke percentage
df["horsepower"].isnull().sum()/ len(df)*100
```

```python
# missing rows ko print krne ka tarika
# df[df.isnull().any(axis=1)]
```

```python
# lets drop missing values from whole dataset
ks.dropna(inplace=True)
ks.isnull().sum()
```

```python
# lets drop missing values from p.column
df.dropna(subset=["horsepower"],inplace=True)
```

```python
# if missing values are more than 50% in a column than its better to drop whole p.column
ks.drop(["p.column"],axis=1)
```

# 4.Split variables for new columns needed

```python
df.head(3)
```

```python
# hm "name" k column ko "car_company" aur "car_name" mn split krne lgy hn
```

```python
#pehle car_company ka column bnayen
df['car_company'] = df['name'].apply(lambda x: x.split()[0])
```

```python
#check
df.head(3)
```

```python
# car_name ka column bnayen
df['car_name'] = df['name'].apply(lambda x: ' '.join(x.split()[1:]))
```

```python
#check
df.head()
```

```python
# ab "name" k column ke zarurat nae drop kr den
df.drop('name', axis=1, inplace=True)
```
